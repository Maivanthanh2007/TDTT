### Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím
Ví dụ: Ta nhập h = 4:
     *
   *   *
 *       *
  * * * * * *  * 


```python
h = int(input("Nhập độ cao h: "))

for i in range(1, h + 1):
    if i == 1:
        print(" " * (h - 1) + "*")
        
    elif i == h:
        print("*" * (2 * h - 1))
        
    else:
        khoang_trang_ngoai = " " * (h - i)
        khoang_trang_trong = " " * (2 * i - 3)
        print(khoang_trang_ngoai + "*" + khoang_trang_trong + "*")
```

### Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 

```python
# Nhập một chuỗi các số cách nhau bởi dấu cách trên cùng 1 dòng
day_so = map(int, input().split())

lon_nhat = None
nho_nhat = None

for n in day_so:
    if n == -1:
        break
        
    if lon_nhat is None or n > lon_nhat:
        lon_nhat = n
    if nho_nhat is None or n < nho_nhat:
        nho_nhat = n

if lon_nhat is not None:
    print(lon_nhat)
    print(nho_nhat)
```

### Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay không 


```python
n = int(input())

if n <= 0:
    print("nhập lại")
else:
    temp = n
    while temp % 3 == 0 and temp > 0:
        temp //= 3
    
    if temp == 1:
        print(n, "là số có dạng 3^k")
    else:
        print(n, "không phải dạng 3^k")
```

### Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A 


```python
A = int(input("Nhập số nguyên dương A: "))

f1 = 1
f2 = 1

if A < 1:
    print("Không có số Fibonacci nào phù hợp")
elif A == 1:
    print("Số Fibonacci lớn nhất là: 1")
else:
    while f1 + f2 <= A:
        f_next = f1 + f2
        f1 = f2
        f2 = f_next
    
    print("Số Fibonacci lớn nhất không vượt quá", A, "là:", f2)
```
