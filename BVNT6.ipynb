### W6A16
Nhâ ̣p vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải


```python
n, m = map(int, input("Nhập số hàng n và số cột m: ").split())

matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

for row in matrix:
    for val in row:
        print(f"{val:>4}", end="")
    print()
```

### W6A17
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
đường chéo phụ của mảng


```python
n = int(input("Nhập kích thước n của ma trận vuông: "))

matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

duong_cheo_chinh = [matrix[i][i] for i in range(n)]

duong_cheo_phu = [matrix[i][n - 1 - i] for i in range(n)]

print("Đường chéo chính:", *duong_cheo_chinh)
print("Đường chéo phụ:", *duong_cheo_phu)
```

### W6A18
Nhâ ̣p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột
k (hàng k) 


```python
n, m, k = map(int, input("Nhập n, m, k: ").split())

matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

tong_hang_k = sum(matrix[k])

tong_cot_k = sum(row[k] for row in matrix)

print(f"Tổng hàng {k}: {tong_hang_k}")
print(f"Tổng cột {k}: {tong_cot_k}")
```

### W6A19
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
trong list

```python
def tim_vi_tri_max(nums):
    val_max = max(nums)
    return nums.index(val_max)

danh_sach = list(map(int, input("Nhập list số nguyên: ").split()))
print("Vị trí có giá trị lớn nhất là:", tim_vi_tri_max(danh_sach))
```

### W6A20
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
-1

```python
def tim_vi_tri_k(nums, k):
    if k in nums:
        return nums.index(k)
    return -1

danh_sach = list(map(int, input("Nhập list số nguyên: ").split()))
k = int(input("Nhập số k cần tìm: "))

print(tim_vi_tri_k(danh_sach, k))
```

### W6A21
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện


```python
strings = input("Nhập các chuỗi: ").split()

pos_dict = {}

for i, s in enumerate(strings):
    if s not in pos_dict:
        pos_dict[s] = []
    pos_dict[s].append(i)

result = {k: tuple(v) for k, v in pos_dict.items()}

print(result)
```

### W6A22
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
trong list


```python
def tim_vi_tri_max(nums):
    if not nums:
        return -1
    
    gia_tri_max = max(nums)
    vi_tri = nums.index(gia_tri_max)
    
    return vi_tri

danh_sach = [10, 5, 20, 15, 20]
print(tim_vi_tri_max(danh_sach))
```

### W6A23
VIết hàm xóa đi các số <x có trong mảng


```python
def xoa_nho_hon_x(nums, x):
    return [n for n in nums if n >= x]

arr = [1, 5, 2, 8, 3, 0]
x = 4
result = xoa_nho_hon_x(arr, x)
print(result)
```

### W6A24
Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9


```python
def day_zero_ve_dau(nums):
    zeros = [x for x in nums if x == 0]
    others = [x for x in nums if x != 0]
    return zeros + others

arr = [0, 2, 3, 2, 4, 0, 0, 9]
print(day_zero_ve_dau(arr))
```

### W6A25
Cho dãy chỉ gồm N giá trị từ 0 - 9 . Kiểm tra xem giá trị nào không có trong
mảng

```python
day_so = input("Nhập dãy số (cách nhau bởi dấu cách): ").split()
da_co = set(int(x) for x in day_so)

tat_ca = set(range(10))
con_thieu = tat_ca - da_co

print("Các giá trị không có trong mảng là:", sorted (list(con_thieu)))
```

### W6A26
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
phần tử đầu và cuối

```python
ds = [int(x) for x in input("Nhập danh sách các số: ").split()]

tong = 0

for i in range(1, len(ds) - 1):
    if ds[i] > ds[i-1] and ds[i] > ds[i+1]:
        tong += ds[i]

print("Tổng các phần tử cực đại địa phương là:", tong)
```

### W6A27
Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước).
nếu không có in ra -1


```python
day_so = [int(i) for i in input("Nhập dãy số: ").split()]
x = int(input("Nhập giá trị x: "))

vi_tri = []

for i in range(len(day_so)):
    if day_so[i] < x:
        vi_tri.append(i)

if len(vi_tri) > 0:
    print("Vị trí các số nhỏ hơn x là:", *vi_tri)
else:
    print(-1)
```

### W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
a = [int(x) for x in input("Nhập dãy số: ").split()]

la_tang = all(a[i] < a[i+1] for i in range(len(a) - 1))
la_giam = all(a[i] > a[i+1] for i in range(len(a) - 1))

if la_tang:
    print("Dãy tăng thực sự")
elif la_giam:
    print("Dãy giảm thực sự")
else:
    print("Vô danh")
```
