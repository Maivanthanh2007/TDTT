Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) ->

```python
def Max_Of_Two(a, b) -> int:
    if a > b:
        return a
    return b

num1 = int(input("Nhập số thứ nhất: "))
num2 = int(input("Nhập số thứ hai: "))

print("Số lớn hơn là:", Max_Of_Two(num1, num2))
```

### W5A2
Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]: 


```python
from typing import Tuple

def swap(a, b) -> Tuple[int, int]:
    a, b = b, a
    return a, b

num1 = int(input("Nhập số thứ nhất: "))
num2 = int(input("Nhập số thứ hai: "))

num1, num2 = swap(num1, num2)

print("Sau khi hoán đổi:")
print("Số thứ nhất:", num1)
print("Số thứ hai:", num2)
```

### W5A3
Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
import math

def is_prime(n) -> bool:
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Nhập số cần kiểm tra: "))

if is_prime(num):
    print(num, "là số nguyên tố")
else:
    print(num, "không phải là số nguyên tố")
```

### W5A4
VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_perfect_number(n) -> bool:
    if n < 1:
        return False
    
    tong_uoc = 0
    for i in range(1, n):
        if n % i == 0:
            tong_uoc += i
            
    return tong_uoc == n

num = int(input("Nhập số cần kiểm tra: "))

if is_perfect_number(num):
    print(num, "là số hoàn hảo")
else:
    print(num, "không phải là số hoàn hảo")
```

### W5A5
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 


```python
def find_index(lst, k) -> int:
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1

danh_sach = [int(x) for x in input("Nhập danh sách số nguyên (cách nhau bởi dấu cách): ").split()]
k = int(input("Nhập số nguyên k cần tìm: "))

print("Vị trí tìm thấy:", find_index(danh_sach, k))
```

### W5A6
Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python
def tinh_giai_thua(n) -> int:
    ket_qua = 1
    for i in range(1, n + 1):
        ket_qua *= i
    return ket_qua

n = int(input("Nhập số nguyên n: "))

if n < 0:
    print("Không tính được giai thừa của số âm")
else:
    print(f"{n}! =", tinh_giai_thua(n))
```

### W5A7
Viết hàm giả lập một máy tính bỏ túi.
Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat
\)
Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
Phép trừ \( - \)
Phép cộng \( + \)
Phép chia \( / \)
Phép nhân \( * \)
Kết quả làm tròn đến chữ số thập phân thứ hai.
Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
Đầu ra: Kết quả phép toán. 


```python
def calculator(num1, num2, operat) -> float:
    if operat == '+':
        res = num1 + num2
    elif operat == '-':
        res = num1 - num2
    elif operat == '*':
        res = num1 * num2
    elif operat == '/':
        if num2 == 0:
            return "Lỗi chia cho 0"
        res = num1 / num2
    else:
        return "Toán tử không hợp lệ"
    
    return round(res, 2)

n1 = float(input("Nhập số thứ nhất: "))
op = input("Nhập phép toán (+, -, *, /): ")
n2 = float(input("Nhập số thứ hai: "))

print("Kết quả:", calculator(n1, n2, op))
```

### W5A8
Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong
một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá
trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn. 


```python
def hamming_distance(x: int, y: int) -> int:
    xor_result = x ^ y
    count = 0
    while xor_result > 0:
        count += xor_result & 1
        xor_result >>= 1
    return count

num1 = int(input("Nhập số x: "))
num2 = int(input("Nhập số y: "))

print("Khoảng cách Hamming:", hamming_distance(num1, num2))
```

### W5A9
Viết hàm trả về tổng các chữ số của số nguyên dương n

```python
def sum_of_digits(n) -> int:
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total

num = int(input("Nhập số nguyên dương n: "))

if num < 0:
    print("Vui lòng nhập số dương")
else:
    print("Tổng các chữ số:", sum_of_digits(num))
```

### W5A10
Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài. 


```python
def is_isomorphic(a, b) -> bool:
    mapping = {}
    mapped_chars = set()
    
    for i in range(len(a)):
        char_a = a[i]
        char_b = b[i]
        
        if char_a in mapping:
            if mapping[char_a] != char_b:
                return False
        else:
            if char_b in mapped_chars:
                return False
            mapping[char_a] = char_b
            mapped_chars.add(char_b)
            
    return True

s1 = input("Nhập chuỗi thứ nhất: ")
s2 = input("Nhập chuỗi thứ hai: ")

print(is_isomorphic(s1, s2))
```

### W5A12
Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|.

```python
from typing import Tuple

def divide_with_remainder(dividend, divisor) -> Tuple[int, int]:
    q = dividend // divisor
    r = dividend % divisor
    return q, r

num1 = int(input("Nhập số bị chia (dividend): "))
num2 = int(input("Nhập số chia (divisor): "))

if num2 == 0:
    print("Lỗi: Không thể chia cho 0")
else:
    q, r = divide_with_remainder(num1, num2)
    print(f"Thương (q) = {q}, Số dư (r) = {r}")
```

### W5A14
Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 


```python
def compute(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        if b == 0:
            return "Lỗi chia cho 0"
        return a / b
    else:
        return "Toán tử không hợp lệ"

so1 = float(input("Nhập số a: "))
toan_tu = input("Nhập phép tính (+, -, *, /): ")
so2 = float(input("Nhập số b: "))

print("Kết quả:", compute(so1, so2, toan_tu))
```

### W5A15
 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle(a, b, c) -> bool:
    if a <= 0 or b <= 0 or c <= 0:
        return False
    
    side1, side2, side3 = sorted([a, b, c])
    return side1**2 + side2**2 == side3**2

val1 = float(input("Nhập cạnh a: "))
val2 = float(input("Nhập cạnh b: "))
val3 = float(input("Nhập cạnh c: "))

print(is_right_triangle(val1, val2, val3))
```
