### W5A16
Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_armstrong(x: int) -> bool:
    if x < 0:
        return False
    
    s = str(x)
    n = len(s)
    tong = sum(int(chu_so) ** n for chu_so in s)
    
    return tong == x

num = int(input("Nhập số cần kiểm tra: "))
print(is_armstrong(num))
```

### W5A17
Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
import math

def is_perfect_square(x: int) -> bool:
    if x < 0:
        return False
    can_bac_hai = math.isqrt(x)
    return can_bac_hai * can_bac_hai == x

num = int(input("Nhập số cần kiểm tra: "))
print(is_perfect_square(num))
```

### W5A19
VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.


```python
def is_magic_number(n: int) -> bool:
    if n <= 0:
        return False
    
    while n >= 10:
        tong = 0
        while n > 0:
            tong += n % 10
            n //= 10
        n = tong
        
    return n == 1

num = int(input("Nhập số cần kiểm tra: "))
print(is_magic_number(num))
```

### W5A20
Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except ValueError:
        try:
            float(s)
            return 'float'
        except ValueError:
            return 'neither'
a=input("Nhập một chuỗi: ")
print(detect_number_type(a))
```

Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

### W5A25
Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1) 

```python
def point_in_quadrilateral(quad, p):
    def cross_product(a, b, c):
        return (b[0] - a[0]) * (c[1] - a[1]) - (b[1] - a[1]) * (c[0] - a[0])

    cp1 = cross_product(quad[0], quad[1], p)
    cp2 = cross_product(quad[1], quad[2], p)
    cp3 = cross_product(quad[2], quad[3], p)
    cp4 = cross_product(quad[3], quad[0], p)

    if (cp1 >= 0 and cp2 >= 0 and cp3 >= 0 and cp4 >= 0) or \
       (cp1 <= 0 and cp2 <= 0 and cp3 <= 0 and cp4 <= 0):
        return True
    return False

quad = [tuple(map(int, input().split())) for _ in range(4)]
p = tuple(map(int, input().split()))
print(point_in_quadrilateral(quad, p))
```

### W5A27
Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def count_non_zero_digits(numbers):
    total_count = 0
    for n in numbers:
        s = str(abs(n))
        for char in s:
            if char != '0':
                total_count += 1
    return total_count

data = input()
num_list = [int(x) for x in data.split()]

result = count_non_zero_digits(num_list)
print("Tổng số lượng chữ số khác 0 là:", result)
```

### W5A28
Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
import math

def calculate_angle_ABC(x1, y1, x2, y2, x3, y3):
    ba_x, ba_y = x1 - x2, y1 - y2
    bc_x, bc_y = x3 - x2, y3 - y2

    dot_product = ba_x * bc_x + ba_y * bc_y
    
    magnitude_ba = math.sqrt(ba_x**2 + ba_y**2)
    magnitude_bc = math.sqrt(bc_x**2 + bc_y**2)

    cos_angle = dot_product / (magnitude_ba * magnitude_bc)
    
    cos_angle = max(-1.0, min(1.0, cos_angle))

    angle_rad = math.acos(cos_angle)
    return math.degrees(angle_rad)

x1, y1 = map(float, input("Nhập tọa độ A (x y): ").split())
x2, y2 = map(float, input("Nhập tọa độ B (x y): ").split())
x3, y3 = map(float, input("Nhập tọa độ C (x y): ").split())

print("Góc ABC bằng:", calculate_angle_ABC(x1, y1, x2, y2, x3, y3), "độ")
```

### W5A33
Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,
giá km, phí chờ, phụ phí ban đêm và giảm giá.


```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4:
        open_price, price_per_km = 12000, 15000
    else:
        open_price, price_per_km = 15000, 18000

    fare = open_price + km * price_per_km
    
    wait_fee = wait_minutes * 1000
    
    night_surcharge = 0
    if start_hour >= 22 or start_hour < 5:
        night_surcharge = fare * 0.2
        
    total = fare + wait_fee + night_surcharge
    
    if km > 30:
        total = total * 0.9
        
    return total

loai_xe = int(input("Nhập loại xe (4 hoặc 7 chỗ): "))
gio_di = int(input("Nhập giờ đi (0-23): "))
so_km = float(input("Nhập số km: "))
phut_cho = int(input("Nhập số phút chờ: "))

print("Tổng tiền taxi:", smart_taxi_fare(loai_xe, gio_di, so_km, phut_cho))
```

### W5A34
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):
    if n <= 0:
        return 0
    
    total = 0
    for x in range(1, n + 1):
        s = str(x)
        if len(s) == len(set(s)):
            total += x
    return total

n_input = int(input("Nhập n: "))
print("Tổng các số thỏa mãn điều kiện là:", sum_1_to_n(n_input))
```
