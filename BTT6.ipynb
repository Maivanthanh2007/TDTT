### W6A1
Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]


```python
data = input("Nhập các số nguyên: ")
numbers = [int(x) for x in data.split()]

result = []
for x in numbers:
    if x not in result:
        result.append(x)

print(result)
```

### W6A2
Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu
tới vị trí i), giữ nguyên thứ tự

```python
data = input("Nhập các số nguyên: ")
numbers = [int(x) for x in data.split()]

result = []
current_sum = 0

for x in numbers:
    current_sum += x
    result.append(current_sum)

print(result)
```

### W6A3
Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple
sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp
xếp.)


```python
data = input("Nhập các số nguyên (cách nhau bởi dấu cách): ")
tup = tuple(int(x) for x in data.split())
k = int(input("Nhập số nguyên k: "))

k = k % len(tup) if len(tup) > 0 else 0

result = tup[k:] + tup[:k]

print(result)
```

### W6A4
Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key
lặp lại, thêm vào list của key đó 

```python
data = input("Nhập các cặp key:value (cách nhau bởi dấu cách): ")
pairs = data.split()

result = {}

for pair in pairs:
    key, value = pair.split(':')
    if key not in result:
        result[key] = [value]
    else:
        result[key].append(value)

print(result)
```

### W6A5
Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ...,
'zeros': ...} (đếm số >0, <0, =0)

```python
nums = list(map(int, input("Nhập các số cách nhau bằng khoảng trắng: ").split()))

result = {'positives': 0, 'negatives': 0, 'zeros': 0}

for n in nums:
    if n > 0:
        result['positives'] += 1
    elif n < 0:
        result['negatives'] += 1
    else:
        result['zeros'] += 1

print(result)
```

### W6A9
Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
theo thứ tự từ điển tăng dần của key.  

```python
dict1 = {'a': 10, 'b': 20, 'c': 30}
dict2 = {'b': 5, 'c': 10, 'd': 15}

result = dict1.copy()

for key, value in dict2.items():
    if key in result:
        result[key] += value
    else:
        result[key] = value

sorted_result = dict(sorted(result.items()))

print(sorted_result)
```

### W6A10
Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

```python
nums = list(map(int, input("Nhập danh sách số nguyên: ").split()))
k = int(input("Nhập số nguyên dương k: "))

pairs = []

n = len(nums)
for i in range(n):
    for j in range(i + 1, n):
        if nums[i] + nums[j] == k:
            pairs.append((nums[i], nums[j]))

result = sorted(pairs, key=lambda x: x[0])

print(result)
```

### W6A11
Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa
tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên
thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt. 

```python
data = tuple(map(int, input("Nhập các số nguyên (cách nhau bởi dấu cách): ").split()))

chan = tuple(x for x in data if x % 2 == 0)
le = tuple(x for x in data if x % 2 != 0)

print(chan)
print(le)
```

### W6A12
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong
list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ
nhất trong số đó

```python
nums = list(map(int, input("Nhập danh sách số nguyên: ").split()))

unique_nums = sorted(set(nums))
result = max(unique_nums, key=nums.count)

print(result)
```

### W6A14
Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các
phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các
phần tử chung đó, không trùng lặp.


```python
list1 = list(map(int, input("Nhập danh sách 1: ").split()))
list2 = list(map(int, input("Nhập danh sách 2: ").split()))

set1 = set(list1)
set2 = set(list2)

giao = list(set1 & set2)

print(giao)
```

### W6A15
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp
key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary
mới.

```python
data = {'apple': 10, 'banana': 5, 'orange': 20, 'grape': 3}
k = int(input("Nhập số nguyên k: "))

result = {key: value for key, value in data.items() if value > k}

print(result)
```
