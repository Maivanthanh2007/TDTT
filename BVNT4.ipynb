### W4A20
Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 


```python
n = int(input())

temp = n
while temp > 1 and temp % 2 == 0:
    temp //= 2

if temp == 1:
    print("YES")
else:
    print("NO")
```

### W4A23

```python
n = int(input())

s = 0
k = 0

while s + (k + 1) < n:
    k += 1
    s += k

print(k)
```

### W4A24
Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 


```python
a = float(input())

n = 0
tong = 0

while tong <= a:
    n += 1
    tong += 1 / n

print(n)
```

### W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 


```python
numbers = []

while True:
    n = int(input())
    if n == -1:
        break
    numbers.append(n)

if numbers:
    print(max(numbers), min(numbers))
```

### W4A26
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 


```python
a = int(input())

f1, f2 = 1, 1

if a < 1:
    print(0)
else:
    while f1 + f2 <= a:
        f_next = f1 + f2
        f1 = f2
        f2 = f_next
    
    print(f2)
```

### W4A27
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng)
Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
s = input()

length = 0
for char in s:
    length += 1

count = 0
for i in range(length):
    if s[i] != " ":
        if i == 0 or s[i-1] == " ":
            count += 1

print(count)
```

### W4A28
Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi
Dùng cách: for trên index; KHÔNG dùng hàm split()  

```python
s = input()

length = 0
for char in s:
    length += 1

start = -1
for i in range(length):
    if s[i] != " ":
        start = i
        break

if start != -1:
    word = ""
    for i in range(start, length):
        if s[i] == " ":
            break
        word += s[i]
    print(word)
```

### W4A29
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành
thường và thường thành hoa

```python
s = input()
ket_qua = ""

for i in range(len(s)):
    char = s[i]
    if 'a' <= char <= 'z':
        ket_qua += chr(ord(char) - 32)
    elif 'A' <= char <= 'Z':
        ket_qua += chr(ord(char) + 32)
    else:
        ket_qua += char

print(ket_qua)
```

### W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số 

```python
s = input()

hoa = 0
thuong = 0
so = 0

for char in s:
    if char.isupper():
        hoa += 1
    elif char.islower():
        thuong += 1
    elif char.isdigit():
        so += 1

print(hoa, thuong, so)
```
