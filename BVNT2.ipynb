### Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
import math
a,b,c=map(float,input().split())
if a+b>c and a+c>b and b+c>a:
    p=(a+b+c)/2    
    print("Diện tích tam giác:", math.sqrt(p*(p-a)*(p-b)*(p-c)))
    print("Chu vi tam giác:",a+b+c)
    if a==b==c:
            print("Đây là tam giác đều")
    elif a==b or b==c or a==c:
            print("Đây là tam giác cân")
    elif a**2==b**2+c**2 or b**2==a**2+c**2 or c**2==a**2+b**2:
            print("Đây là tam giác vuông")
    else:
            print("Đây là tam giác thường")
else:
    print("Không phải là tam giác")
        
```

### Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math
a,b,c=map(float,input().split())
delta= b**2-4*a*c
if delta<0:
    print("Phương trình vo nghiệm")
elif delta==0:
    print("Phương trình có nghiệm kép là x=", -b/(2*a))
else:
    print("Phương trình có hai nghiệm phân biêt là x1=", (-b+math.sqrt(delta))/(2*a), "và x2=", (-b-math.sqrt(delta))/(2*a))
```

### Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
a,b=map(float,input().split())
r=float(input())
x,y=map(float,input().split())
if (x-a)**2 + (y-b)**2==r**2:
    print("Điểm nằm trên đường tròn")
elif (x-a)**2 + (y-b)**2<r**2:
    print("Điểm nằm trong đường tròn")
else:   
    print("Điểm nằm ngoài đường tròn")
```
